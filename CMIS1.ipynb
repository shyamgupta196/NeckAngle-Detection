```python
#1
def uniquechassis(chassis):   #to check whether chassis is unique or not
    f = open('.cars.txt','r')
    for j in f.readlines():
        a = j.split(',')
        if a[0] == chassis:
            return True
def addcar():    #add details to cars.txt
    chassis = input("Enter chassis number: ")
    while uniquechassis(chassis) :
        print('This chassis already exists!! ')
        chassis = input("Enter chassis number Again: ")
    while len(chassis)!=17:  #check whether chassis have 17 characters or not
        print('enter the chassis of length 17')
        chassis = input("Enter chassis number Again: ")
    make = input("Enter making company: ")
    model_year = input('enter model year: ')
    body_type = input("Enter Body type:")
    buying_price = input("Enter Buying Price:")
    milage = input("Enter milage:")
    Type = input("Enter Type:")
    quantity = input("Enter Quantity:")
    
    data = (chassis, make,model_year,body_type,buying_price,milage,quantity,Type)
    f = open('cars.txt', 'a')
    f.write('\n')
    f.write(','.join(data))
```

```python
#2
def uniquechassis(chassis):     #to check chassis is unique or not
    f = open('sales.txt','r')
    for j in f.readlines():
        a = j.split(',')
        if a[3] == chassis:
            return True
def sales():         #to add details to sales.txt
    sales = open('sales.txt', 'a')
    cars = open('cars.txt','a+')
    mainten = open('maintenance.txt', 'r')
    invoice_no = input('Enter invoice_no: ')
    date = input('Enter date: ')
    customer_id = input("Enter customer ID:")
    customer_name = input('Enter customer_name: ')
    chassis = input('Enter car_chassis_no: ')
    while uniquechassis(chassis) :
        print('This chassis already exists!! ')
        chassis = input("Enter chassis number Again: ")
    while len(chassis)!=17:   #to check whether chassis is of length 17 or not
        print('enter the chassis of length 17')
        chassis = input("Enter chassis number Again: ")
    Type = input('Enter Type: ')
    quantity = input('Enter quantity: ')
    for i in cars.readlines():
        a = i.split(',')
        if a[0] == chassis:
            if a[-2] == 0 or a[-2] < quantity:
                print("The required quantity is not present as of now.")
                print("Please, look for another car!")
                sell()
            cars.write(i.replace(a[-2],str(int(a[-2])-int(quantity))))   #updates quantity in cars.txt
    cars.close()
    sales_price = input('Enter sales_price:')
    customer_mobile_no = input('Enter customer_mobile_no: ')
    while len(customer_mobile_no) != 10:
        print('Mobile number should have 10 digits!')
        customer_mobile_no = input('Enter customer_mobile_no: ')
    email = input('Enter email: ')
    
    cars = open('cars.txt','r')  #To access buying price
    for j in cars.readlines():
        b = j.split(',')
        if b[0] == chassis:
            buying_price = int(b[-4])
    for i in mainten.readlines():  #To access maintenance price
        a = i.split(',')
        if a[2] == chassis:
            maintenance_price = int(a[-2])
            
    profit = int(sales_price) - maintenance_price - buying_price    #Calculates profit earned
    data = (invoice_no,date,customer_name,chassis,Type,quantity,sales_price,str(profit),customer_mobile_no,email)
    sales.write('\n')
    sales.write(','.join(data))
    sales.close()
    cars.close()
    mainten.close()
```

```python
#3
def editdetails():  #To edit details in cars.txt
    f = open('cars.txt','r')
    lines = f.readlines()
    chasis = input('Enter chasis number: ')
    for i in lines:
        a = i.split(',')
        if a[0]==chasis:
            print(i)
            index = int(input('which index column do you want to change: '))
            old = a[index]
            new = input('enter new value: ')
            a[index] = new
            replacement = i.replace(old,new)
            lines.pop(lines.index(i))
            lines.append(replacement)
    f.close()
    f = open('cars.txt','w')
    for i in lines:
        f.writelines(i)
    f.close()
```

```python
#4
def maintenance():  #To enter details in maintenance.txt
    maintenance_id = input("Enter maintanance ID:")
    date = input("Enter date:")
    chassis_no = input("Enter chassis number:")
    cost_price = input("Enter cost price:")
    maintenance_type = input("Enter cost price:")
    
    data = (maintenance_id,date,chassis_no,cost_price,maintenance_type)
    
    f = open('maintenance.txt','w')
    f.write(','.join(data))
```

```python
# 5
def cars_available():   #To display cars available for sales
    f = open('cars.txt','r')
    l = f.readlines()
    for j in l[1:]:
        
        if int(j.split(',')[-2])>0:
            print(j)
    f.close()
```

```python
#6
def display():  #To display information about a specific car
    f = open('cars.txt','r')
    chassis = input("Enter chassis number:")
    l = f.readlines()
    print(l[0])
    for j in l:
        a = j.split(',')
        if a[0] == chassis:
            view = " ".join(map(str,a))
            print(view)
    f.close()
```

```python
#7
def displaymake():  #To display car of a certain make
    make = input("Enter make company:")
    f = open('cars.txt','r')
    l = f.readlines()
    header = l[0]
    new = []
    for i in l:
        a = i.split(',')
        if a[1] == make:
            new.append(a)
    for i in range(len(new)):
        for j in range(0, len(new) - i - 1):
            if int(new[j][4]) > int(new[j+1][4]):
                new[j], new[j+1] = new[j+1],new[j]
    print(header)
    for i in new:
        print(' '.join(i))
```

```python
#8
def monthlyreport(): #To display monthly report
    sale = open('sales.txt','r')
    car  = open('cars.txt','r')

    sale_head = sale.readlines()
    car_head = car.readlines()

    months = set()
    for i in sale_head[1:]:
        if len(i) !=0:
            a = i.split(',')
            date = a[1].split('/')
            months.add(date[1])

    for j in months:
        monthly_sales= 0
        monthly_revenue = 0
        total_profit = 0

        for i in sale_head[1:]:
            a = i.split(',')
            if j in a[1]:
                monthly_sales += int(a[7])
                total_profit += int(a[8])
                monthly_revenue +=(int(a[6])*int(a[7]))
        print(f'monthly sales for {j} month --> {monthly_sales}')
        print(f'monthly revenue for {j} month --> {monthly_revenue}')
        av_daily_sale = monthly_sales/30
        print(f'average daily sales for {j} month-->{av_daily_sale}')
        profit = total_profit/30
        print(f'profit from sales for {j} month-->{profit}')

    month = int(input(f'from given choices select a month {months}'))
    for i in j:
        cars_sold = 0
        if month == i:
            for k in sale_head[1:]:
                a = i.split(',')
                cars_sold += int(a[6])
            print(cars_sold)
            
    #since month is not specified for cars.txt, we display total number of cars
    cars = open('cars.txt','r')
    cars_head = cars.readlines()
    total_cars = 0
    for i in cars_head[1:]:
        a = i.split(',')
        total_cars += int(a[-2])
    print(total_cars)
```

```python
def main(): #contains all the function combined
    while True:
        print("""
1. Add a car
2. Sell a car
3. Edit car details
4. Add maintenance details
5. Display available cars for sale
6. Display information about a specific car
7. Display cars of a certain brand
8. Monthly report
9. Exit
""")
        ans = int(input("Enter number:"))
        if ans == 1:
            addcar()
        elif ans == 2:
            sales()
        elif ans == 3:
            editdetails()
        elif ans == 4:
            maintenance()
        elif ans == 5:
            cars_available()
        elif ans == 6:
            display()
        elif ans == 7:
            displaymake()
        elif ans == 8:
            monthlyreport()
        elif ans == 9:
            break
        else:
            print("Wrong input! Enter again.")
main()
```
